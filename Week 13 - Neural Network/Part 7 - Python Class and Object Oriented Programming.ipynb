# Class

```python
class Manusia:
    def __init__(self, warna_rambut, tinggi, berat):
        self.warna_rambut = warna_rambut
        self.tinggi = tinggi
        self.berat = berat
        
    def cat_rambutnya(self, warna_baru):
        self.warna_rambut = warna_baru
        print(f"Rambut saya jadi warna {warna_baru}")
        
    def BMI(self):
        return self.berat / self.tinggi ** 2
```

```python
wira = Manusia("hitam", 170, 60)
desti = Manusia("hitam", 155, 45)
```

```python
wira.BMI()
```

```python
desti.BMI()
```

```python
class CustomNN(nn.Module):
    def __init__(self, input_size, output_size, dropout=0.2):
        super().__init__()        
        self.fc = nn.Sequential(
            nn.Linear(input_size, 16),
            nn.ReLU(),
            nn.Dropout(dropout),

            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Dropout(dropout),

            nn.Linear(8, output_size),
            nn.LogSoftmax(1)    
        )
        
    def forward(self, x):
        return self.fc(x)
```
